```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.__version__
pd.__version__
```

```python
df_sem1 = pd.read_csv("Downloads/MOCK_DATA_Sem1.csv")
df_sem1.tail(10)
```

```python
df_sem1["Total"] = df_sem1["PhysicsI"]+df_sem1["ChemistryI"]+df_sem1["MathematicsI"]+df_sem1["BiologyI"]
df_sem1.tail()
```

```python
df_sem1["Status"] = None
df_sem1.tail()
```

```python
for index,physics, chemistry, biology, mathematics in zip(df_sem1.index,df_sem1["PhysicsI"],df_sem1["ChemistryI"],df_sem1["BiologyI"],df_sem1["MathematicsI"]):
    if physics <= 30 or chemistry <= 30 or biology <= 30 or mathematics <= 30 :
        df_sem1.at[index,'Status']='Fail'
        df_sem1.at[index,'Color']='red'
    else :
        df_sem1.at[index,'Status']='Pass'
        df_sem1.at[index,'Color']='green'
```

```python
df_sem1.head()
```

```python
data = [[df_sem1.columns[1], len(df_sem1[df_sem1[df_sem1.columns[1]]>=30]), len(df_sem1[df_sem1[df_sem1.columns[1]]<30])],
       [df_sem1.columns[2], len(df_sem1[df_sem1[df_sem1.columns[2]]>=30]), len(df_sem1[df_sem1[df_sem1.columns[2]]<30])],
       [df_sem1.columns[3], len(df_sem1[df_sem1[df_sem1.columns[3]]>=30]), len(df_sem1[df_sem1[df_sem1.columns[3]]<30])],
        [df_sem1.columns[4], len(df_sem1[df_sem1[df_sem1.columns[4]]>=30]), len(df_sem1[df_sem1[df_sem1.columns[4]]<30])]
       ]
df_subject = pd.DataFrame(data,columns= ['Subjects', 'Pass','Fail'])
df_subject = df_subject.set_index('Subjects')
```

```python
y = [len(df_sem1[df_sem1['Status'] == 'Pass']),len(df_sem1[df_sem1['Status'] == 'Fail'])]
mylabels = ["Pass","Fail"]
mycolors = ['green', 'red']
plt.title("Pass and Fail Students in Semester 1")
plt.pie(y, labels = mylabels, colors = mycolors, autopct='%1.1f%%', explode=(0.1, 0.1))

plt.legend()
plt.show()
```
